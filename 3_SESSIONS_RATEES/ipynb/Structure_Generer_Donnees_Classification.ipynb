## Presentation

Tour des **generateurs sklearn** pour creer des jeux de donnees synthetiques (classification, regression, clustering). Indispensable pour : tester un modele, illustrer un concept, reproduire un exemple, evaluer la robustesse a differents types de difficultes (non-lineaire, desequilibre, bruit, multi-classe).

Chaque generateur teste un type de difficulte specifique. Le notebook visualise les frontieres de decision typiques.

## Plan

1. `make_classification` (parametrique generique)
2. `make_blobs` (clusters gaussiens)
3. `make_moons` (non-lineaire)
4. `make_circles` (non-lineaire, isotrope)
5. `make_gaussian_quantiles` (multi-classe concentrique)
6. Generation multi-classe et desequilibre
7. Tableau : quel generateur teste quel concept
8. References


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import (
    make_classification, make_blobs, make_moons, make_circles,
    make_gaussian_quantiles, make_regression,
)
SEED = 42
np.random.seed(SEED)
```

## 1. `make_classification` â€” generique parametrique

Le plus flexible : controle de `n_features`, `n_informative`, `n_redundant`, `class_sep`, `flip_y` (bruit).

```python
X, y = make_classification(
    n_samples=500, n_features=2, n_informative=2, n_redundant=0,
    n_classes=2, n_clusters_per_class=1,
    class_sep=1.0, flip_y=0.05, random_state=SEED,
)
print(f"X shape={X.shape}, y unique={np.unique(y, return_counts=True)}")

plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k", s=40)
plt.title("make_classification (class_sep=1.0)")
plt.show()
```

## 2. `make_blobs` â€” clusters gaussiens isotropes

Lineairement separable. Test ideal des baselines (LogReg, kNN, SVM lineaire).

```python
X, y = make_blobs(n_samples=500, centers=3, cluster_std=1.0, random_state=SEED)
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="tab10", edgecolors="k", s=40)
plt.title("make_blobs (3 centres)")
plt.show()
```

## 3. `make_moons` â€” non-lineaire

2 demi-lunes imbriquees. Necessite un modele non-lineaire (kernel SVM, RF, MLP).

```python
X, y = make_moons(n_samples=500, noise=0.15, random_state=SEED)
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k", s=40)
plt.title("make_moons (noise=0.15)")
plt.show()
```

## 4. `make_circles` â€” non-lineaire isotrope

2 cercles concentriques. Pas separable lineairement. Necessite kernel ou feature engineering radial.

```python
X, y = make_circles(n_samples=500, factor=0.5, noise=0.1, random_state=SEED)
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k", s=40)
plt.title("make_circles (factor=0.5)")
plt.show()
```

## 5. `make_gaussian_quantiles` â€” multi-classe concentrique

```python
X, y = make_gaussian_quantiles(n_samples=500, n_features=2, n_classes=4, random_state=SEED)
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="tab10", edgecolors="k", s=40)
plt.title("make_gaussian_quantiles (4 classes)")
plt.show()
```

## 6. Desequilibre â€” `weights` argument

```python
X, y = make_classification(
    n_samples=1000, n_features=2, n_informative=2, n_redundant=0,
    n_clusters_per_class=1, weights=[0.9, 0.1],   # 90/10
    random_state=SEED,
)
unique, counts = np.unique(y, return_counts=True)
print(f"Distribution classes : {dict(zip(unique, counts))}")

plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k", s=40)
plt.title("make_classification desequilibre 90/10")
plt.show()
```

## 7. Recap : quel generateur pour quel concept ?

| Generateur | Difficulte | Modeles qui marchent | Modeles qui galerent |
|---|---|---|---|
| `make_blobs` | Lineaire separable | Tous | Aucun (trivial) |
| `make_classification` | Parametrique | Selon `class_sep`, `flip_y` | Selon param |
| `make_moons` | Non-lineaire | SVM RBF, RF, MLP, kNN | LogReg, SVM lineaire |
| `make_circles` | Non-lineaire isotrope | SVM RBF, RF, MLP | LogReg, SVM lineaire |
| `make_gaussian_quantiles` | Multi-classe concentrique | RF, MLP | LogReg lineaire |
| `make_classification(weights=[0.9, 0.1])` | Desequilibre | + class_weight / SMOTE | Sans tuning |
| `make_regression` | Regression generique | LinReg / GBM | - |
